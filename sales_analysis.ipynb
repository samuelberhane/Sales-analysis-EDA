# Exploratory Data Analysis
## Sales Analysis

#### Import python libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os
import calendar
```

#### Merge dataset into single file

```python
dataframe = []
files = [file for file in os.listdir(os.path.join("./Sales_Data"))]
for file in files:
    df = pd.read_csv(os.path.join(f"./Sales_Data/{file}"))
    dataframe.append(df)

sales_data = pd.concat(dataframe)
sales_data.to_csv("sales_data.csv", index=False)
```

```python
df = pd.read_csv("sales_data.csv")
df
```

#### Clean date(drop NaN values)

```python
df = df.dropna(axis=0)
df
```

#### Add Month column to dataframe

```python
df["Month"] = df["Order Date"].str[0:2]
```

```python
df
```

#### Drop rows with Month column contain "Or"

```python
df = df[df["Month"].str.contains("Or")==False]
```

```python
df["Month"] = df["Month"].apply(int)
```

```python
df
```

```python
df["Month"].value_counts().sort_index().plot(kind="bar", title="Amount of sales per month")
```

#### Change numeric column of string to numeric values

```python
df["Quantity Ordered"] = pd.to_numeric(df["Quantity Ordered"])
df["Price Each"] = pd.to_numeric(df["Price Each"])
```

```python
df["Total Sales"] = df["Quantity Ordered"] * df["Price Each"]
```

```python
df.insert(4, 'Total Sales', df.pop('Total Sales'))
```

```python
df
```

```python
sales_per_month = df.groupby(by=["Month"]).sum()
sales_per_month
```

```python
X = pd.Series(sales_per_month.index)
str_month = X.apply(lambda x: calendar.month_name[x][0:3])
Y = sales_per_month["Total Sales"]
plt.plot(str_month,Y)
plt.title("Total Sales For Each Month")
plt.ylabel("Total Sales (USD)")
plt.xlabel("Month of the year")
```

### 2.  In which city is the most profit generated?

#### Add city column

```python
df["City"] = df["Purchase Address"].str.split(",").apply(lambda x: x[1])
```

```python
sales_per_city = df.groupby(by=["City"]).sum()
sales_per_city
```

```python
X = sales_per_city.index
Y = sales_per_city["Total Sales"]
plt.bar(X,Y)
plt.xticks(rotation=90)
plt.title("Total Sales Per City")
plt.ylabel("Total Sales (USD)")
plt.xlabel("City")
```
